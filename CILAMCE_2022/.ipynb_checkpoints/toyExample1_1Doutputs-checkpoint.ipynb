In this notebook, we seek to reproduce the toy example proposed in the paper:

- Andrés F. López-Lopera, Franck Massa, Isabelle Turpin and Nicolas Leconte, "Modeling complex mechanical computer codes with functional input via Gaussian processes", XLIII Ibero-Latin American Congress on Computational Methods in Engineering (CILAMCE), 2022.

First, we load some packages and define useful functions:

```r
library(kergp)       # modified kergp package (install file kergp_0.5.0.tar)
require(VGAM)        # package for Rayleigh distributions
require(ggplot2)     # package for graphics and visualisations
require(repr)        # package for graphics and visualisations

rm(list = ls())      # cleaning global environment

## local functions to account for functional data ##
source("rscripts/corkernelfi.R")
source("rscripts/basisCompute.R")
source("rscripts/futils.R")

## defining a seed ##
seed <- 123 # runif(1, 0, 100) # set a seed for reproducibility
set.seed(seed)
```

## Dataset

- We first define some functions for creating the dataset. 

```r
## sampling from a Gaussian distribution using the Cholesky matrix ##
mvrnormChol <- function(n = 1, mu = rep(0, nrow(L)), L) {
  d <- nrow(L)
  v <- matrix(rnorm(d*n), d, n)
  x <- t(L) %*% v + matrix(mu, d, n)
  return(t(x))
}

## 1D Matern 5/2 kernel ##
kMatern52 <- function(t1, t2 = t1, par) {
  # t1, t2 = vectors with time instants
  # par = (variance, lengthscale)
  sigma2 <- par[1]; theta <- par[2]
  
  # computing the kernel
  dist2 <- outer(t1[, 1]/theta, t2[, 1]/theta, '-')^2
  sqrt5dist <- sqrt(5)*sqrt(dist2)
  expSqrt5dist <- exp(-sqrt5dist)
  return(sigma2*(1 + sqrt5dist + (5/3)*dist2)*expSqrt5dist)
}

## 1D Squared Exponential kernel ##
kSE <- function(t1, t2 = t1, par) {
  # t1, t2 = vectors with time instants
  # par = (variance, lengthscale)
  sigma2 <- par[1]; theta <- par[2]
  
  # computing the kernel
  dist2 <- outer(t1[, 1]/theta, t2[, 1]/theta, '-')^2
  return(sigma2*exp(-dist2))
}
```

- We generate the synthetic functional inputs used in this notebook. We sample 101 functional scenarios: the first 200 are used for training the Gaussian process model and the last one is used for testing. The number of training scenarios (``nrepTrain``) can be increased for a better quality of predictions. **Warning:** The computational cost of the Gaussian process model increases as the ``nrepTrain`` increases! We recommend to slightly increase ``nrepTrain`` (e.g. 150, 200, ...) for further testing.

```r
#### building up the toy dataset ####
## sampling random functional profiles ##
nrepTrain <- 100 # nb of functional replicates for training
nrepTest <- 1 # nb of functional replicates for testing
nrep <- nrepTrain + nrepTest

# functional data #
nu <- 1e2 # input evaluations
u <- as.matrix(seq(0, 1, length.out = nu)) # independent variable of the functional input
dimF <- 3 # dimension of the functional input space

fu <- vector("list", dimF) # list containing the functional inputs per scenario
scale_ray <- seq(0.05, 1, length.out = dimF) # interval for the shape parameter of the Rayleigh distribution 
par(mfrow = c(1,3))
for (i in 1:dimF) {
    fu_iter <- matrix(NaN, nrep, nu)
    for (j in 1:nrep) {
        seed <- seed + 1
        set.seed(seed)
        fu_iter[j, ] <- drayleigh(u, scale = runif(1, min(scale_ray), max(scale_ray)))
        fu_iter[j, ] <-  runif(1, 2, 4)*fu_iter[j, ]/max(fu_iter[j, ]) # modified version of the Rayleigh distribution (see the paper)
    }
    fu[[i]] <- fu_iter
    options(repr.plot.width = 20, repr.plot.height = 5)
    matplot(u, t(fu[[i]]), type = "l", xlab = "u", ylab = "f(u)")
    
}
```

- We generate the synthetic functional outputs used in this notebook. We sample random realization from a Gaussian process using the functional input generated previously. To do so, we need to define some parameters (e.g. type of the kernel function).

```r
## parameters associated to the output (time-series) ##
nb_t <- 50 # nb of instant points
t <- matrix(seq(0, 1, length.out = nb_t), ncol = 1)
kernType_t <- k1FunMatern5_2 # kernel type (e.g. k1FunMatern5_2, kSE)
par_t <- c(0.04, 1) # kernel parameters
names(par_t) <- c("theta", "sigma2")

## parameters associated to the functional inputs ##
projType <- "PCA" # type of the projection onto basis
f <- basisCompute(fu, projType, nu) # PCA decomposition on the functional inputs
attr(f, "projType") <- projType
attr(f, "kernType") <- kernType_f <- "Matern52" # kernel type
groupList <- f # list with functional spaces (F1, ..., Fq)
par_f <- rep(20, length.out = dimF) # kernel parameters
names(par_f) <- c(paste("theta_f", 1:dimF, sep = ""))

## defining kernels to be used by kergp ##
kt <- covMan(kernel = kernType_t,
             inputs =  "t",
             hasGrad = TRUE,
             d = 1L,
             par = par_t,
             parLower = c(theta = 0.0, sigma2 = 0.0),
             parUpper = c(theta = Inf, sigma2 = Inf),
             parNames = c("theta", "sigma2"),
)

kf <- covFun(kernel = kfCompute,
             inputs =  "u",
             groupList = groupList, # list with functional spaces (F1, ..., Fq)
             hasGrad = TRUE,
             label = "functional kernel",
             par = par_f,
             parLower = rep(0.01, length(par_f)),
             parUpper = rep(Inf, length(par_f)),
             parNames = eval(parse(text = paste("c(", 
                                                paste("'theta_f", 1:dimF, "'", sep = "", collapse = ", "), ")")))
)
```

- We now sample random Gaussian process trajectories via Cholesky decomposition.

```r
## sampling random time-series ##
ubase <- 1:nrep # sequence used as index for the scenarios
Lf <- chol(covMat(object = kf, X = ubase, Xnew = ubase))
Lt <- chol(covMat(object = kt, X = t, Xnew = t))
L <- Lf %x% Lt
set.seed(seed)
ysamp <- matrix(mvrnormChol(1, rep(0, nrow(L)), L), nrow = nrow(t))
options(repr.plot.width = 10, repr.plot.height = 5)
matplot(ysamp, type = "l", xlab = "t", ylab = "F(t)")
```

## Gaussian process model with functional inputs

- Next, we train and test the resulting GP model. For the PCA representation of the functional data, we consider a total inertia of $95\%$. We precompute the distances between functional replicates looking for computational savings.

```r
#### predicting some data ####

## defining initial parameters ##
# temporal kernel #
par_t <- c(0.1, 1)
names(par_t) <- c("theta", "sigma2")
kt@par <- par_t

# functional kernel #
par_f <- rep(10, dimF)
names(par_f) <- paste("theta_f", 1:dimF, sep = "")
kf@par <- par_f

## building up the list with functional data via PCA ##
# list with training data #
nbasis <- NULL
funDataTrain <- lapply(fu, "[", -nrep, )
ftrain <- basisCompute(funDataTrain, projType, lv.inertia = 0.999)

# list with test data given the training data #
funDataTest <- lapply(fu, "[", nrep, )
ftest <- PCABasisComputeNew(funDataTest, ftrain)
groupListTest <- ftest

# distance between functional replicates are precomputed for computational savings
distffTest <- dist2FFvec(groupListTest$coef, groupListTest$coef, groupListTest$GramM)
groupListTest$distff <- distffTest
kf@groupList <- groupListTest

# defining the separable kernel #
kmixKro <- covComp(formula = ~ kf() %x% kt())
```

- Training the Gaussian process model using the function ``gp``.

```r
## training the model ##
ysamp <- lapply(1:nrep, function(i) ysamp[ , i]) # list to be passed to the model
ydesign <- unlist(ysamp[1:nrepTrain]) # using the first nrep-1 scenarios to train the model
Xdesign <- data.frame(t = t,
                      u = rep(ubase[1:nrepTrain], each = nrow(t)))
myData <- Xdesign
myData$Y <- ydesign # dataframe containing the training data

## testing cpu.time of the model using the Kronecker product ##
time2GPKro <- proc.time()
modelOptKro <- gp(formula = Y ~ 1, cov = kmixKro,
                  data = myData, inputs = c('u', 't'),
                  noise = FALSE, estim = TRUE, compGrad = FALSE,
                  trace = 0)
time2GPKro <- proc.time() - time2GPKro
```

- Computing the Gaussian process predicting using the function ``predict``. 

```r
## predicting the last scenario (time-series) ##
t_test <- seq(0, 1, length.out = nb_t)
x <- expand.grid(t = t_test, u = ubase[nrep])
time2predKro <- proc.time()
predOptKro <- predict(modelOptKro, x, type = 'SK')
time2predKro <- proc.time() - time2predKro
```

- Plotting the Gaussian process prediction with 3-sd conficende intervals.

```r
nbstd <- 3
df <- data.frame(t = c(t, t_test),
                 f = c(ysamp[[nrep]], predOptKro$mean),
                 fmin = c(NULL*ysamp[[nrep]], predOptKro$mean-nbstd*predOptKro$sd),
                 fmax = c(NULL*ysamp[[nrep]], predOptKro$mean+nbstd*predOptKro$sd),
                 label = c(rep("target function", length(t)), rep("GP mean", length(t_test))))

options(repr.plot.width = 6, repr.plot.height = 5)
ggplot(data=df, aes(t, f)) +
    geom_line(mapping = aes(x = t, y = f, colour = label),
              linetype = c(rep("solid", length(t)), rep("longdash", length(t))), size = 1) +
    geom_ribbon(aes(x = t, ymin = fmin, ymax = fmax, group = label),
                colour ="gray90", fill = "gray60", alpha = 0.1) +
    theme_bw() + theme(legend.position="bottom") +
    scale_x_continuous(name = "t") + scale_y_continuous(name = "F(t)")
```
